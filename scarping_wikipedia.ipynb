```python
page = requests.get('https://en.wikipedia.org/wiki/Car')
soup = BeautifulSoup(page.content, "html.parser")
```

```python
af = soup.find_all('table', class_="multicol")
for result in af:
    results = result.find_all("a")
for result in results:
    if str(result['href']).startswith('/wiki/'):
        print("https://en.wikipedia.org" + result['href'])
```

```python
all_links = []
for lin in wiki_links:
    page = requests.get(lin)
    soup = BeautifulSoup(page.content, "html.parser")
    af = soup.find_all('table', class_="multicol")
    for result in af:
        results = result.find_all("a")
    for result in results:
        if str(result['href']).startswith('/wiki/'):
            all_links.append("https://en.wikipedia.org" + result['href'])
            print("https://en.wikipedia.org" + result['href'])
```

```python
web_text = ""
for el in set_links:
    if 'en.wikipedia' in el:
        page = requests.get(el)
        soup = BeautifulSoup(page.content, "html.parser")
        # getting all the paragraphs
        for para in soup.find_all("p"):
            web_text = web_text + para.get_text() + "\n"
    #         print(para.get_text())
```
