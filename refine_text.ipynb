```python
new_file = open("word_file/nn_car.txt", "w")
for line in lines:
    contains_digit = any(map(str.isdigit, line))
    if not contains_digit:
        new_file.write(line)
new_file.close()
```

```python
a_file = open("OneLook.txt", "r")
lines = a_file.readlines()
a_file.close()

new_file = open("OneLook.txt", "w")
for line in lines:
    contains_digit = any(map(str.isdigit, line))
    if contains_digit:
        _, r = line.split('. ')
        new_file.write(r)
new_file.close()
```

```python
txt_files = glob.glob("word_file/*.txt")
txt_files
```

```python
all_words = []
for file in txt_files:
    a_file = open(file, "r")
    lines = a_file.readlines()
    lines = [el.strip().lower() for el in lines]
    all_words.extend(lines)
    a_file.close()
all_words = set(all_words)
```

```python
from nltk.corpus import stopwords
print(stopwords.words('english'))
```

```python
filtered_words = [word for word in all_words if word not in stopwords.words('english') and ' ' not in word]
filtered_words
```

```python
len(filtered_words)
```

```python
not_stemed_filtered_words = [el.replace(" ", "_") for el in filtered_words if ' ' not in el]
```

```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()
```

```python
stemed_filtered_words = set([porter.stem(el).replace(" ", "_") for el in filtered_words])
stemed_filtered_words = [el.replace(" ", "_") for el in stemed_filtered_words]
stemed_filtered_words
```

```python
import gensim.downloader
print(list(gensim.downloader.info()['models'].keys()))
```

```python
glove_vectors = gensim.downloader.load('glove-wiki-gigaword-300')
```

```python
! ls ~/gensim-data
```

```python
data = []
for w in not_stemed_filtered_words:
    if w in glove_vectors.index_to_key:
        sim_val = glove_vectors.similarity("car", w)
        print(f'similarity score for {w} is: {sim_val}')
    else:
        print(f'{w} not presented in the gensim model')
        sim_val = np.nan
    data.append([w, sim_val])
    print('\n __________________________________________ \n')
```

```python
not_stemed_words_df = pd.DataFrame(data, columns = ['word', 'similarity'])
pd.set_option('display.max_rows', None)
not_stemed_words_df = not_stemed_words_df.sort_values('similarity', ascending=False, ignore_index=True)
not_stemed_words_df = not_stemed_words_df.iloc[1: , :]
not_stemed_words_df
```

```python
not_stemed_words_df.to_csv('not_stemmed_word_based_wiki_score.csv', index=False)
```
